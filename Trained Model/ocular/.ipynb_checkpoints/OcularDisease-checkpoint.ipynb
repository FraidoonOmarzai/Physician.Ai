**We just need to extract normal and cataract images**

```python
def has_cataract(text):
    if "cataract" in text:
        return 1
    else:
        return 0
```

```python
df["left_cataract"] = df["Left-Diagnostic Keywords"].apply(lambda x: has_cataract(x))
df["right_cataract"] = df["Right-Diagnostic Keywords"].apply(lambda x: has_cataract(x))
df.head()
```

```python
left_cataract_imgs = []
right_cataract_imgs = []
for i in range(len(df)):
    if df["left_cataract"][i] == 1:
        left_cataract_imgs.append(df['Left-Fundus'][i])
    if df["right_cataract"][i] == 1:
        right_cataract_imgs.append(df['Right-Fundus'][i])
```

```python
def is_normal(text):
    if "normal fundus" in text:
        return 1
    else:
        return 0
```

```python
df["left_normal"] = df["Left-Diagnostic Keywords"].apply(lambda x: is_normal(x))
df["right_normal"] = df["Right-Diagnostic Keywords"].apply(lambda x: is_normal(x))
```

```python
left_normal_imgs = []
right_normal_imgs = []
for i in range(len(df)):
    if df["left_normal"][i] == 1:
        left_normal_imgs.append(df['Left-Fundus'][i])
    if df["right_normal"][i] == 1:
        right_normal_imgs.append(df['Right-Fundus'][i])
```

```python
print(f'left_cataract_imgs: {len(left_cataract_imgs)}')
print(f'right_cataract_imgs: {len(right_cataract_imgs)}')
print(f'total_cataract_imgs: {len(left_cataract_imgs) + len(right_cataract_imgs)}')
print(20*"-")
print(f'left_noramal_imgs: {len(left_normal_imgs)}')
print(f'right_normal_imgs: {len(right_normal_imgs)}')
print(f'total_normal_imgs: {len(left_normal_imgs) + len(right_normal_imgs)}')
```

**balance the data**

```python
left_normal = df.loc[(df.C ==0) & (df["Left-Diagnostic Keywords"] == "normal fundus")]["Left-Fundus"].sample(300,random_state=42).values
right_normal = df.loc[(df.C ==0) & (df["Right-Diagnostic Keywords"] == "normal fundus")]["Right-Fundus"].sample(300,random_state=42).values
print(f'left_noramal_imgs: {len(left_normal)}')
print(f'right_normal_imgs: {len(right_normal)}')
```

**dataset creation**

```python
cataract = np.concatenate((left_cataract_imgs,right_cataract_imgs),axis=0)
normal = np.concatenate((left_normal,right_normal),axis=0)
```

```python
cataract[0]
```

```python
from tensorflow.keras.preprocessing.image import load_img,img_to_array
from tqdm import tqdm

dataset_dir = "preprocessed_images/"
image_size=224
labels = []
dataset = []

def create_dataset(image_category,label):
    for img in tqdm(image_category):
        image_path = os.path.join(dataset_dir,img)
        try:
            image = cv2.imread(image_path,cv2.IMREAD_COLOR)
            image = cv2.resize(image,(image_size,image_size))
        except:
            continue
        
        dataset.append([np.array(image),np.array(label)])
    # random.shuffle(dataset)
    return dataset
```

```python
dataset = create_dataset(cataract,1)
dataset = create_dataset(normal,0)
```

```python
dataset
```

```python
x = np.array([i[0] for i in dataset]).reshape(-1,image_size,image_size,3)
y = np.array([i[1] for i in dataset])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

## Train the model

```python
base_model = tf.keras.applications.EfficientNetB0(include_top=False)
base_model.trainable = False

inputs = layers.Input(shape=(224, 224, 3), name="input_layer") 
x = base_model(inputs, training=False) 
x = layers.GlobalAveragePooling2D(name="global_average_pooling")(x)
x = layers.Dense(128, activation='relu')(x)
x = Dropout(0.2)(x)
x = layers.Dense(64, activation='relu')(x)
x = layers.Dense(32, activation='relu')(x)
outputs = layers.Dense(1, activation="sigmoid", name="output_layer")(x)
model = tf.keras.Model(inputs, outputs)
```

```python
model.summary()
```

```python
# Compile
model.compile(loss="binary_crossentropy",
              optimizer=tf.keras.optimizers.Adam(),
              metrics=['accuracy'])
```

```python
# Fit
history = model.fit(x_train,
                    y_train,
                    epochs=10,
                    validation_data=(x_test,y_test)
                    )
```

```python
history
```

```python
pd.DataFrame(history.history).plot()
```

```python
model.evaluate(x_test, y_test)
```

**save the trained model**

```python
model.save("model.h5")
```
